```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('titles.csv')
df = df.dropna()

# # remove punctionation (tag = Z)
df = df[df.Tags != "Z"]
# df = df.set_index("Tags")
# df = df.drop("Z", axis=0) 
```

```python
df.Lemma.unique()
df.Lemma.value_counts() 
```

```python
df["isVerb"] = df["Tags"].str.find('V') 
verbs = df[df.isVerb >= 0]
verbs
```

```python
verbs["isMain"] = verbs["Tags"].str.find('Vm') 
main_verbs = verbs[verbs.isMain >= 0]
main_verbs
```

```python
sd = main_verbs.Lemma.value_counts()
```

```python
plt.plot(main_verbs.Lemma.value_counts())
plt.ylabel('some numbers')
plt.show()
```

```python
sd = pd.DataFrame(sd)
sd = sd[sd.Lemma > 1]
```

```python
names = sd.index
values = sd["Lemma"]


plt.barh(names, values)
```
